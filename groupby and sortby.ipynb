# Groupby

```python
import pandas as pd
```

```python
drinks=pd.read_csv('D:/JOB/PYTHON/lecture_practice/groupby/drinks.csv')
```

```python
drinks.head(2)
```

```python
##average beer servings across all countries.

drinks.beer_servings.mean()
```

```python
##if we have to categorise beer_servings across the continents

drinks.groupby('continent').beer_servings.mean()
```

```python
## how groupby works

drinks[drinks.continent=='Asia'].beer_servings.mean()
```

```python
drinks[drinks.continent=='Africa'].beer_servings.mean()
```

```python
drinks[drinks.continent=='Europe'].beer_servings.mean()
```

```python
drinks[drinks.continent=='North America'].beer_servings.mean()
```

```python
drinks[drinks.continent=='South America'].beer_servings.mean()
```

```python
##specify multiple aggregate functions

drinks.groupby('continent').beer_servings.agg(['count','mean','max','min'])
```

```python
##if we just use drinks with groupby then all the categories will be separated

drinks.groupby('continent').mean()
```

Groupby should be used for analysis of some panda series by some category. for the above example the category 'Continent'. if in the query for each is coming then we use groupby.
